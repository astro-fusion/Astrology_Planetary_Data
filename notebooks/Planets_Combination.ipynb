# Vedic Astrology Data

```python
import pandas as pd
import datetime
```

```python
all_dfs = []  # List to store DataFrames for each year

for year in range(1600, 2100):
    url = f'https://raw.githubusercontent.com/astro-fusion/Astrology_Planetary_Data/refs/heads/main/data/planetary_positions_simplified/{year}.csv'
    df = pd.read_csv(url)
    df['year'] = year  # Add a 'year' column to each DataFrame
    all_dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(all_dfs, ignore_index=True)
```

## Data wrangling

### Subtask:
Transform the DataFrame `df` to create a new column named 'planets_per_house'.  This column will contain dictionaries where keys are house numbers (1 through 12) and values are lists of planets located in that house.


**Reasoning**:
Create the 'planets_per_house' column by iterating through the DataFrame and creating dictionaries for each row, then append the dictionary to a list and assign the list to the new column.



```python
planets_per_house_list = []
for index, row in df.iterrows():
    planets_dict = {}
    for i in range(1, 13):
        house_column = f'House {i}'
        planets_in_house = []
        if pd.notna(row[house_column]):
            planets = row[house_column].split()
            planets_in_house.extend(planets)
        planets_dict[i] = planets_in_house
    planets_per_house_list.append(planets_dict)

df['planets_per_house'] = planets_per_house_list
display(df.head())
```

## Data analysis

### Subtask:
Count the number of combinations where more than 4 planets are in the same house.


**Reasoning**:
Iterate through the DataFrame and count combinations where more than 4 planets are in the same house.



```python
count = 0
for index, row in df.iterrows():
    planets_per_house_dict = row['planets_per_house']
    for house_number, planets in planets_per_house_dict.items():
        if len(planets) >= 5:
            count += 1
print(f"The number of combinations where more than 4 planets are in the same house is: {count}")
```

```python
year = 1600  # The year of your data
base_date = datetime.date(year, 1, 1)

approximate_dates = []
for row_index in combination_rows:
    date = base_date + datetime.timedelta(days=row_index)
    approximate_dates.append(date)

print("Approximate dates of combinations:")
for date in approximate_dates:
    print(date)
```

```python
dates = approximate_dates
combined_dates = []
i = 0
while i < len(dates):
    current_date = dates[i]
    j = i + 1
    while j < len(dates) and (dates[j] - current_date).days <= 3:
        j += 1
    combined_dates.append(min(dates[i:j]) + datetime.timedelta(days=(max(dates[i:j]) - min(dates[i:j])).days // 2)) #using midpoint date
    i = j

print("Combined dates of combinations:")
for date in combined_dates:
    print(date)
```
